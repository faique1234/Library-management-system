```python
from random import randint
from getpass import getpass
members = []
member=    {"CNIC": "12345", "Name": "Faique", "Book Issued": 0}
members.append(member)

books = []
book={"Serial no": "101", "Name": "Python Basics", "Author": "John Doe", "Status": "Available"}
books.append(book)
status = "Available"
# Book-related class
class bookdetails:
    def bookmenu(self):
      while True:
        print("|Select Your Choice |")
        print("|1. Add book        |")
        print("|2. Search book     |")
        print("|3. Delete book     |")
        choice = input("Enter your choice: ").strip()
        if choice == "1":
            self.addbook()
            break
        elif choice == "2":
            self.searchbook()
            break
        elif choice == "3":
            self.deletebook()
            break
        else:
            print("Invalid choice")


    def addbook(self):
        name = input("| Enter the name of book: ").strip().title()
        author = input("| Enter the name of author: ").strip().title()
        status = "Available"
        serial = randint(1000, 9999)
        book = {"Name": name, "Author": author , "Status":status , "Serial no": serial}
        books.append(book)
        print(f"Book Added: Serial no: {book['Serial no']} \t Name: {book['Name']} \t Author: {book['Author']} \t Status {book['Status']}")
        self.back_to_menu()

    def searchbook(self):
        name = input("Enter the book name to search: ").strip().title()
        for book in books:
            if book["Name"] == name:
                print(f"Book Found: Serial no: {book['Serial no']} \t Name: {book['Name']} \t Author: {book['Author']} \t Status {book['Status']}")

                break
        else:
            print("Book not found.")
        self.back_to_menu()

    def deletebook(self):
        name = input("Enter the name of book to delete: ").strip().title()
        for book in books:
            if book["Name"] == name:
                books.remove(book)
                print("Book deleted successfully.")
                break
        else:
            print("Book not found.")
        self.back_to_menu()

    def back_to_menu(self):
      while True:
        choice = input("Enter 'm' to return to menu or 'e' to exit: ").strip().lower()
        if choice == 'm':
            self.menu()
        elif choice == 'e':
            main()

        else:
            print("Invalid choice")



# Member-related class
class memberdetails:
    def membermenu(self):
      while True:
        print("| Select Your Choice|")
        print("| 1. Add member     |")
        print("| 2. Search member  |")
        print("| 3. Delete member  |")
        print("| 4. Back to menu   |")
        choice = input("Enter your choice: ").strip()
        if choice == "1":
            self.addmember()
            break
        elif choice == "2":
            self.searchmember()
            break
        elif choice == "3":
            self.deletemember()
            break
        elif choice == "4":
            self.back_to_menu()
            break
        else:
            print("Invalid choice")

    def addmember(self):
        name = input("| Enter member name: ").strip().title()
        cnic = input("| Enter member CNIC [CNIC will be used as ID]: ").strip()
        book_issued = 0
        member = {"Name": name, "CNIC": cnic,"Book Issued":book_issued}
        members.append(member)
        print(f"Member Added: Name: {member['Name']} \t ID: {member['CNIC']} \t Book Issued: {member['Book Issued']}")
        self.back_to_menu()

    def searchmember(self):
      while True:

        cnic = input("Enter member ID to search: ").strip()
        for member in members:
            if member["CNIC"] == cnic:
                print(f"Member Found: Name: {member['Name']} \t ID: {member['CNIC']} \t Book Issued: {member['Book Issued']}")

                break
            elif member["CNIC"] != cnic:
                print("Member not found.")
                break
            else:
                print("Invalid choice")
                break
        self.back_to_menu()

    def deletemember(self):
        cnic = input("Enter member ID to delete: ").strip()
        for member in members:
            if member["CNIC"] == cnic:
                members.remove(member)
                print("Member deleted successfully.")
                self.back_to_menu()
                break
            else:
                print("Member not found.")
                self.back_to_menu()
                break

    def back_to_menu(self):
      while True:
        choice = input("Enter 'm' to return to menu or 'e' to exit: ").strip().lower()
        if choice == 'm':
            self.menu()
            break
        elif choice == 'e':
            main()
        else:
            print("invalid  Entry")



# Main Menu class
class menuselect(bookdetails, memberdetails):
    def __init__(self):
        print("Welcome to Library System")

    def menu(self):
        print("| Main Menu:        |")
        print("| 1. Book Details   |")
        print("| 2. Member Details |")
        print("| 3. Log Out        |")
        choice = input("Enter your choice: ").strip()
        if choice == "1":
            self.bookmenu()
        elif choice == "2":
            self.membermenu()
        elif choice == "3":
            main()
        else:
            print("Invalid choice")
            self.menu()


# Admin login inherits from menu (which inherits everything else)
class adminlogin(menuselect):
    def __init__(self):
        super().__init__()

    def admin_login(self):
        print("---- Admin Login ----")
        user_id = input("Enter your ID: ").strip().title()
        password = getpass("Enter your password: ").strip()
        if user_id == "Faique" and password == "12345":
            print("Access granted. Welcome admin!")
            self.menu()
        else:
            print("Access denied. Try again.")
            self.admin_login()



def return_to_menu(menu_obj):
    while True:
        _ = input("Press 'm' for MENU or Press 'e' for EXIT: ").lower().strip()
        if _ == "m":
            menu_obj.menumember()
            break
        elif _ == "e":
          main()
        else:
            print("Wrong Entry")


class MemberMenu:
    def __init__(self):
        print("Welcome to the Library System")
        print("-----------------------------")

    def menumember(self):
        while True:
            print("\n| --- Main Menu ---|")
            print("| 1. Search book   |")
            print("| 2. Borrow book   |")
            print("| 3. Return book   |")
            print("| 4. Log Out       |")
            choice = input("Enter your choice: ").strip()
            if choice == "1":
                self.search()
                break
            elif choice == "2":
                self.book_status("Borrowed")
                break
            elif choice == "3":
                self.book_status("Available")
                break
            elif choice == "4":
                main()
                break
            else:
                print("Wrong Entry")

    def search(self):
        while True:
            print("\n| --- Book Search ---     |")
            print("| 1. Search by book name  |")
            print("| 2. Search by author name|")
            print("| 3. Back to main menu    |")
            choice = input("Enter your choice: ").strip()

            if choice == "1":
                name = input("Enter book name: ").strip().title()
                found = False
                for book in books:
                    if book['Name'] == name:
                        print(f"\nBook Found:\nSerial: {book['Serial no']}, Name: {book['Name']}, Author: {book['Author']}, Status: {book['Status']}")
                        found = True
                        break
                if not found:
                    print("Book not found.")
                return_to_menu(self)
                break

            elif choice == "2":
                author = input("Enter author name: ").strip().title()
                found = False
                for book in books:
                    if book['Author']== author:
                        print(f"\nBook Found:\nSerial: {book['Serial no']}, Name: {book['Name']}, Author: {book['Author']}, Status: {book['Status']}")
                        found = True
                        break
                if not found:
                    print("Book not found.")
                return_to_menu(self)
                break

            elif choice == "3":
                self.menumember()
                break

            else:
                print("Wrong Entry")

    def book_status(self, action):
        serial = input("Enter the serial number of the book: ").strip()
        for book in books:
            if book["Serial no"] == serial:
                if action == "Borrowed":
                    if book["Status"] == "Available":
                        book["Status"] = "Borrowed"
                        print("Book has been borrowed.")
                        for member in members:
                            if member["Book Issued"] == 0:
                                member["Book Issued"] = serial
                                break
                    else:
                        print("Book is already borrowed.")
                elif action == "Available":
                    if book["Status"] == "Borrowed":
                        book["Status"] = "Available"
                        print("Book has been returned.")
                        for member in members:
                            if member["Book Issued"] == serial:
                                member["Book Issued"] = 0
                                break
                    else:
                        print("Book is already available.")
                return_to_menu(self)
                return
        print("Book with this serial number not found.")
        return_to_menu(self)


class MemberLogin(MemberMenu):
    def __init__(self):
        super().__init__()

    def member_login(self):
        print("\n--- Member Login ---")
        print("---- ---- ----")
        name = input("Enter your name: ").strip().title()
        cnic = getpass("Enter your ID: ").strip()
        for member in members:
            if member["CNIC"] == cnic and member['Name']==name:
                print("ACCESS GRANTED. Welcome", member["Name"])
                self.menumember()
                return
        print("ACCESS DENIED. CNIC not found.")

                                             # Main Function
def main ():
 print("Welcome to library management system")
 print("                                                        | LOGIN DASHBOARD  |")
 print("                                                        | 1. Admin Login   |")
 print("                                                        | 2. Member Login  |")
 while True:
  _ = input("Enter your choice ").strip()
  match _:
   case "1":
    a1 = adminlogin()
    a1.admin_login()
    break
   case "2":
    p=MemberLogin()
    p.member_login()
    break
   case _:
    print("Wrong Entry")


main()
```
